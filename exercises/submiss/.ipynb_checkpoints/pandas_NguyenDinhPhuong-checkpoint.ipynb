```python
#1.16 what is most expensive item ordered
#note: \D matches any non-decimal digit
#1st way
# name.item_price = name.item_price.str.replace('\$','').astype('float')
#2nd way
name.item_price.replace(to_replace='\$',value='',inplace=True)#item_price type is string
name.sort_values('item_price',ascending=False).head(1)
```
