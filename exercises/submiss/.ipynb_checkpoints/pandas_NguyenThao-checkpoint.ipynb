```python
data=pd.read_csv('chipotle.tsv',sep="\t")
```

```python
data.head()
```

```python
data[['item_price','item_name']]
```

```python
data['item_name'].value_counts()['Nantucket Nectar']
```

```python
data['item_price_float'] = data['item_price'].replace('\$', '', regex=True).astype(float)
data['price_per_unit'] = data['item_price_float']/data['quantity']
```

```python
sorted_data = data.sort_values(by='price_per_unit',ascending=False)
sorted_data['item_name'].iloc[0]
```

```python
sorted_data[:7]
```
