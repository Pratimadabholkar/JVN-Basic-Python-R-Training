## Show the price of each item

```python
# your code goes here
# Show the avg price of each item.
df['item_price_float'] = df.item_price.apply(lambda t: float(t.replace("$", "")))
df.groupby("item_name").item_price_float.mean()[:10]
```

## How many times were a "Nantucket Nectar" ordered?

```python
# your code goes here
len(df[df.item_name == "Nantucket Nectar"])
```

## What was the most expensive item ordered?

```python
# your code goes here
df[df.item_price_float == df.item_price_float.max()]
```

## Select the first 7 rows of the most expensive item ordered

```python
# your code goes here
df.sort_values("item_price_float", ascending=False).head(7)
```
