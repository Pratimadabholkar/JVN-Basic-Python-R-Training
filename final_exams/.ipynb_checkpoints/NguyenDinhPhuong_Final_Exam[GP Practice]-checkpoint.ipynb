```python
import pandas as pd
import missingno
import re
%matplotlib inline
```

```python
df = pd.read_csv(r'GP-Practices.csv',low_memory=False)
```

```python
df.dropna(axis=0, how='all', inplace=True)
```

```python
df.shape
```

```python
#1b. Cleaning and Standardizing “Practice Type” (e.g. 17J , 2C ).
df['Practice Type'].replace(to_replace=' .+',value='',regex=True, inplace=True)
```

```python
df['Practice Type'].unique()
```

```python
#1c. Formatting the “Postcode” as “XXX-XXX” (e.g. EH5-3AH).
def standardlize_postcode(string):
    temp = re.match('([0-9a-zA-Z]{3,}) *-*([0-9a-zA-Z]{3,})', string)
    return temp[1]+'-'+temp[2]
    
df.Postcode = df.Postcode.apply(standardlize_postcode)
```

```python
df.Postcode.unique()
```

```python
#1d. Cleaning and Standardizing “Public” column (Y = Yes, N = No). Filling all missing value by “N”.
df.Public.fillna('N',inplace=True)
df.Public.replace('Yes','Y',inplace=True)
```

```python
df.Public.unique()
```

```python
#1e. Rename the column “Practice Code” to “PCode”, and “Practice List Size” to “PSize”.
df = df.rename(columns = {'Practice Code':'PCode','Practice List Size':'PSize'})
```

```python
df.dtypes
```

```python
#1f. Replace any negative values in “PSize” with the column average.
_mean = df.PSize.mean()
df.loc[df.PSize < 0,'PSize'] = _mean
```

```python
#1g. In the “Practice Name” column, standardise the strings so that only the first
#letter is uppercase (e.g. "LEVEN MEDICAL PRACTICE" should become "Leven Medical Practice".)
df['Practice Name'] = df['Practice Name'].apply(lambda x: x.title())
```

```python
#1h. The “Month_Year” column would be better as two separate columns! Split each string on the underscore delimiter _ to give 
# two new columns with the correct values.
df1 = pd.DataFrame(df.Month_Year.str.split('_',1).tolist(), columns = ['Month','Year'])
#df.Month_Year.str.split('_',1) is a series, not a list like str.split()
# df1.Month.unique()
df = df.join(df1)
```

```python
#2a. Select the data in rows [3, 4, 8, 20-30] and in columns [“Practice Name”,“Telephone Number”, “Practice Type”, “Public”]
temp = [3,4,8]
temp.extend(range(20,30))
df.loc[temp,['Practice Name','Telephone Number','Practice Type','Public']]
```

```python
#2b. Figure out distribution of “CHP Code” (count values for each “CHP Code”).
df['CHP Code'].value_counts().plot(kind='bar')
```

```python
pd.DataFrame(df['CHP Code'].value_counts()).plot.pie(subplots=True,figsize=(6, 6),autopct='%1.1f%%')
```

```python
#2c. List all unique “Postcode” values having “PSize” > 10000.
df[df.PSize > 10000]['Postcode'].unique()
```

```python
#2d. Calculate the mean “Practice List Size” for each different “CHP Name” in dataframe.
df.groupby('CHP Name')['PSize'].mean()
```

```python
#2e. Sort dataframe first by the values in the “PSize” in descending order, then by the
# value in the “PCode” column in ascending order.
df.sort_values(['PSize','PCode'],ascending=[False,True])
```

```python
#2f. Select the rows that “Postcode” start with “EH3”
df[df.Postcode.str.startswith('EH3')]
```

```python
#2g. Save cleaned file to data_final.xls
df.to_excel('data_final.xls')
```
