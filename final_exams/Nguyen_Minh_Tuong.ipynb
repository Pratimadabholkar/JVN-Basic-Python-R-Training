```python
import pandas as pd
import numpy as np
df = pd.read_csv("GP-Practices.csv")
```

### 1a) Remove these collumns: Telephone Number, CHP Name, CHP Code, Practice Code

```python
del df['Telephone Number']
del df['CHP Name']
del df['CHP Code']
del df['Practice Code']
```

### 1b) Cleaning  and  Standardizing  “Practice  Type” (e.g.  17J ,  2C).

```python
df['Practice Type'].unique()
```

```python
def cleanPT(things):
    temp = things.split()
    result = str(temp[0])
    return result
df['Practice Type']  = df['Practice Type'].apply(cleanPT)
```

```python
df['Practice Type'].unique()
```

### 1c) Formatting  the  “Postcode” as “XXX-XXX” (e.g.  EH5-3AH).

```python
print(df['Postcode'].unique())
```

```python
def cleanPC(things):
    if things.find('-') != -1: # Nếu tìm thấy - thì để yên
        things = things
        return things
    if things.find(' ') != -1: # Nếu không thấy khoảng trắng thì thêm - 
        try:
            temp = things.split()
            result = temp[0]+'-'+temp[1]
        except IndexError:
            print(temp)
        return result
    if things == 'EH91AG':
        things = 'EH9-1AG'
        return things
    if things == 'EH177AB':
        things = 'EH17-7AB'
        return things
    if things == 'EH104RP':
        things = 'EH10-4RP'
        return things
    
                
df['Postcode']  = df['Postcode'].apply(cleanPC)

```

### 1d) Cleaning  and  Standardizing  “Public” column  (Y  =  Yes,  N =  No).  Filling  all missing  value  by “N”.

```python
df['Public'].unique()
```

```python
   
df.Public=df.Public.fillna("N")
df.Public=df.Public.map({'Y':"Y", 'N':"N", 'Yes':"Y"})
```

### 1e) Rename  the  column  “Practice  Code” to  “PCode”,  and  “Practice  List Size” to “PSize”.

```python
df.rename(columns = {'Practice Code':'PCode',"Practice List Size":"PSize"},inplace=True)
```

```python
list(df)
```

### 1f)  Replace  any negative  values in  “PSize ” with  the  column  average.

```python
col_avg = df['PSize'].mean()
u  = df['PSize']
u[u<0] = col_avg

```

### 1g) In  the  “Practice  Name” column,  standardise  the  strings so  that  only the  first letter is uppercase  (e.g.  " LEVEN MEDICAL  PRACTICE "  should  become " Leven Medical  Practice ".)

```python
def process_PT(things):
    things=things.lower()
    things=things.title()
    return things

df["Practice Name"] = df["Practice Name"].apply(process_PT) 
```

### 1h)  The  “Month_Year” column  would  be  better as two  separate  columns!  Split  each string  on  the  underscore  delimiter _  to  give  two  new columns with  the  correct values.

```python
def toMonth(things):
    return things.split("_")[0]

def toYear(things):
    return things.split("_")[-1]

df['Month_Year']=df['Month_Year'].astype(str)
df['Month'] = df['Month_Year'].apply(toMonth)
df['Year'] = df['Month_Year'].apply(toYear)
```
